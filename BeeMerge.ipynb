```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import os
import re
```

```python
weather = "CleanData/Weather_ByPeriod.csv"
bee_count = "CleanData/BeeCount.csv"

weather_df = pd.read_csv(weather, encoding='utf8')
bee_df = pd.read_csv(bee_count, encoding='utf8')
```

```python
weather_df
```

```python
bee_df
```

```python
bee_merged = pd.merge(weather_df, bee_df, how='left', on=['Year','Period'])
bee_merged
```

```python
bee_merged = bee_merged.dropna()

bee_merged
```

```python
bee_merged = bee_merged.drop(['Unnamed: 0'], axis =1)
bee_merged.sort_values(["Year", "Period"]).reset_index(drop=True)
bee_merged
```

```python
bee_merged.to_csv(r'CleanData/BeeMerged.csv')
```
